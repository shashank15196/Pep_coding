```python
### MOBILE KEYPAD

# 1. You are given a string str. The string str will contains numbers only, where each number stands for a key pressed on a mobile phone.
# 2. The following list is the key to characters map
#     0 -> .;
#    1 -> abc
#    2 -> def
#    3 -> ghi
#    4 -> jkl
#    5 -> mno
#    6 -> pqrs
#    7 -> tu
#    8 -> vwx
#    9 -> yz
# 3. Complete the body of printKPC function - without changing signature - to print the list of all words that could be produced by the keys in str.
# Use sample input and output to take idea about output.

# Note -> The online judge can't force you to write the function recursively but that is what the spirit of question is. Write recursive and not iterative logic. The purpose of the question is to aid learning recursion and not test you. 


s=input()
keypad= [" .;", "abc", "def", "ghi", "jkl", "mno", "pqrs", "tu","vwx" ]


def getkeypad(s,ans):
    global keypad
    s=s.replace("\r","")
    if len(s)==0:
        print(ans)
        return
    
    ck=s[0]
    #key= [" .;", "abc", "def", "ghi", "jkl", "mno", "pqrs", "tu","vwx" ]
    for j in keypad[int(ck)]:
        getkeypad(s[1:], ans+j)
        
getkeypad(s,"")
```

```python
#PRINT PERMUTATION


# 1. You are given a string str.
# 2. Complete the body of printPermutations function - without changing signature - to calculate and print all permutations of str.
# Use sample input and output to take idea about permutations.

# Note -> The online judge can't force you to write the function recursively but that is what the spirit of question is. Write recursive and not iterative logic. The purpose of the question is to aid learning recursion and not testyou.

s=input("enter the number: ")

def printpermutation(s,ans):
    
    if(len(s)==0):
        print(ans)
        return
    
    ##choices  ->(abc) -> a,b,c
    
    for i in range(len(s)):
        choice=s[i]
        newinput= s[0:i] + s[i+1:]
        
        printpermutation(newinput,ans+choice)
        
        
printpermutation(s," ")
```
